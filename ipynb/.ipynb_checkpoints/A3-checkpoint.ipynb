```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Amit', 'Sonal', 'Riya', 'Karan', 'Neha', 'Raj'],
    
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    
    'Age': [20, 21, 19, 22, 20, 23],
    
    'Income': [25000, 30000, 28000, 35000, 32000, 40000]
}

df = pd.DataFrame(data)

print(df)
```

```python
grouped = df.groupby('Gender')['Income']
```

```python
print("Mean: ")
print(grouped.mean())
```

```python
print("Median: ")
print(grouped.median())
```

```python
print("Min: ")
print(grouped.min())
```

```python
print("Max: ")
print(grouped.max())
```

```python
print("Standard Deviation: ")
print(grouped.std())
```

```python
male_income = df[df['Gender'] == 'Male']['Income'].tolist()
female_income = df[df['Gender'] == 'Female']['Income'].tolist()
print('Male Income: ', male_income)
print('Female Income: ', female_income)
```
